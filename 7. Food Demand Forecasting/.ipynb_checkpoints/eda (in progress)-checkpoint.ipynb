```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import qgrid

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set(style="ticks", rc=custom_params)

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 20
```

```python
def show_table(df):
    return qgrid.show_grid(df, show_toolbar = True)
```

```python
train_df = pd.read_csv("Datasets/train.csv")
```

```python
store_df = pd.read_csv("Datasets/fulfilment_center_info.csv")
```

```python
meal_df = pd.read_csv("Datasets/meal_info.csv")
```

```python
train_df
```

```python
meal_df
```

```python
store_df
```

```python
df = train_df.merge(store_df, how = "left", on = "center_id").merge(meal_df, how = "left", on = "meal_id")
```

```python
df
```

```python
del train_df, store_df, meal_df
```

```python
df[["center_id", "meal_id", "city_code", "region_code"]] = \
df[["center_id", "meal_id", "city_code", "region_code"]].astype("category")
```

```python
weekly_order_count = df.groupby("week", as_index = False).size()
weekly_order_count["size_diff1"] = weekly_order_count["size"].diff()
```

```python
to_text = weekly_order_count.nsmallest(n = 2, columns = "size_diff1").drop("size_diff1", axis = 1)

plt.figure(figsize = (12,6));

ax = sns.lineplot(x = "week", y = "size", data = weekly_order_count);

weekly_order_count.nsmallest(n = 2, columns = "size_diff1").\
plot(x = "week", y = "size", kind = "scatter", c = "red", ax = ax, legend = True, label = "Lowest Orders")

for i in range(len(to_text)):
    x = to_text.iloc[i, 0]
    y = to_text.iloc[i, 1]
    ax.annotate(f"Week {x} had {y:,} placed orders", (x, y), fontsize=15, va = "bottom")
    
weekly_order_count.set_index("week")["size"].rolling(4).mean().plot(ax = ax, 
                label = "Monthiy average Orders", legend = True)

plt.title("Number of orders by Week");
```

```python
fig, ax = plt.subplots(1, 2, figsize = (13, 7), sharex = True, tight_layout = True)

sns.histplot(x = "checkout_price", data = df, kde = True, ax = ax[0]);
sns.histplot(x = "base_price", data = df, kde = True, ax = ax[1]);
```

```python
fig, ax = plt.subplots(1, 2, figsize = (13, 7), sharex = True, tight_layout = True)

sns.countplot(x = "emailer_for_promotion", data = df, ax = ax[0], edgecolor='k');
sns.countplot(x = "homepage_featured", data = df, ax = ax[1], edgecolor='k');
```

```python
week_order_num = df.groupby("week").agg({"num_orders" : "sum"})

ax = week_order_num.plot(figsize = (14,7))

week_order_num.nsmallest(n = 2, columns = "num_orders").reset_index().\
plot(x = "week", y = "num_orders", kind = "scatter", c = "red", ax = ax, legend = True, 
     label = "Lowest Orders");

for i in range(len(week_order_num.nsmallest(n = 2, columns = "num_orders").reset_index())):
    x = week_order_num.nsmallest(n = 2, columns = "num_orders").reset_index().iloc[i, 0]
    y = week_order_num.nsmallest(n = 2, columns = "num_orders").reset_index().iloc[i, 1]
    ax.annotate(f"Week {x} had {y:,} Orders", (x, y), fontsize=15, va = "bottom")
    
week_order_num.nlargest(n = 2, columns = "num_orders").reset_index().\
plot(x = "week", y = "num_orders", kind = "scatter", c = "green", ax = ax, legend = True, 
     label = "Highest Orders");

for i in range(len(week_order_num.nlargest(n = 2, columns = "num_orders").reset_index())):
    x = week_order_num.nlargest(n = 2, columns = "num_orders").reset_index().iloc[i, 0]
    y = week_order_num.nlargest(n = 2, columns = "num_orders").reset_index().iloc[i, 1]
    ax.annotate(f"Week {x} had {y:,} Orders", (x, y), fontsize=15, va = "bottom")
       
ax.set_title("Total Weekly Number of Orders");
```

```python
df.groupby("center_type").agg({"num_orders" : "sum"}).plot(kind = "bar", rot = 0, figsize = (8, 6), 
                                                           edgecolor='k');
plt.title("Center Type Count")
plt.ticklabel_format(style='plain', axis='y');
```

```python
idx = df.groupby("category").agg({"num_orders" : "sum"}).sort_values("num_orders").index

ax = df.groupby("category").agg({"num_orders" : "sum"}).sort_values("num_orders").\
plot(kind = "bar", rot = 30, figsize = (10, 6), align="center", edgecolor='k');

ax.set_title("Food Category Count")
ax.ticklabel_format(style='plain', axis='y');
ax.set_xticklabels(labels = idx, ha='right');
```

```python
idx = df.groupby("cuisine").agg({"num_orders" : "sum"}).sort_values("num_orders").index

ax = df.groupby("cuisine").agg({"num_orders" : "sum"}).sort_values("num_orders").\
plot(kind = "bar", rot = 0, figsize = (8, 6), align="center", edgecolor='k');

ax.set_title("Cuisine Count")
ax.ticklabel_format(style='plain', axis='y');
```

```python
food_df = df.groupby(["cuisine", "category"]).agg({"num_orders" : "sum"})
tot_cuisine = df.groupby("cuisine").agg({"num_orders" : "sum"})

food_df_orders = food_df.div(tot_cuisine, level = "cuisine").reset_index()
```

```python
plt.figure(figsize = (10, 5))

sns.histplot(food_df_orders, y='cuisine', hue='category', weights='num_orders',
             multiple='stack', palette='tab20c', shrink=0.8, edgecolor='k').\
get_legend().set_bbox_to_anchor((1, 1))

plt.title("Proportion of Food Category Orders by Cuisine");
```
