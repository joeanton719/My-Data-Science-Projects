```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import qgrid

sns.set()

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 20
```

```python
def show_table(df):
    return qgrid.show_grid(df, show_toolbar = True)
```

```python
train_df = pd.read_csv("Datasets/train.csv")
```

```python
store_df = pd.read_csv("Datasets/fulfilment_center_info.csv")
```

```python
meal_df = pd.read_csv("Datasets/meal_info.csv")
```

```python
df = train_df.merge(store_df, how = "left", on = "center_id").merge(meal_df, how = "left", on = "meal_id")
```

```python
df
```

```python
del train_df, store_df, meal_df
```

```python
weekly_orders = df.groupby("week", as_index = False).size()
```

```python
weekly_orders["size_diff1"] = weekly_orders["size"].diff()
```

```python
to_text = weekly_orders.nsmallest(n = 2, columns = "size_diff1").drop("size_diff1", axis = 1)

plt.figure(figsize = (12,6));

ax = sns.lineplot(x = "week", y = "size", data = weekly_orders);

weekly_orders.nsmallest(n = 2, columns = "size_diff1").\
plot(x = "week", y = "size", kind = "scatter", c = "red", ax = ax, legend = True, label = "Lowest Orders")

for i in range(len(to_text)):
    x = to_text.iloc[i, 0]
    y = to_text.iloc[i, 1]
    ax.annotate(f"Week {x}, {y} Orders", (x, y), fontsize=15, va = "bottom")
    
weekly_orders.set_index("week")["size"].rolling(4).mean().plot(ax = ax, 
                label = "Monthiy average Orders", legend = True)

plt.title("Number of orders by Week");
```
